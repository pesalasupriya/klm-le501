```python
year=1900
if (year%400==0 or (year%100!=0 and year%4==0)):
    print(" leap year")
else:
    print("non leap year")
```

```python
def isleapyear(year):
    if (year%400==0 or (year%100!=0 and year%4==0)):
        return True
isleapyear(2000)
```

```python
for i in range(2000,2023):
    if isleapyear(i):
        print(i,end=" ")
```

```python
def isprime(n):
    c=0
    for i in range(1,n+1):
        if n%1==0:
            c=c+1
    if c==2:
        return True
    else:
        return False
isprime(22)
```

```python
def isperfect(n):
    s=0
    for i in range(1,n):
        if n%i==0:
            s=s+i
    if n==s:
        return True
    else:
        return False
isperfect(6)
```

```python
for i in range(1,100001):
    if isperfect(i):
        print(i,end=" ")
```
